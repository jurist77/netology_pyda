```python
import Levenshtein
```

```python
claim_list = [
    'Шехавцова Анна',
    'Гречихина Наталья',
    'Козлова Алена',
    'Груздева Алина',
    'Кущенко Анна',
    'Чистякова Анастасия'
]
```

```python
speech_recognition = [
    'кучменко она',
    'кущенко оксана',
    'груздь алина',
    'рычихина наталья',
    'шиховцева на',
    'чистова анастасия'
]
```

```python
for i in speech_recognition:
    for j in claim_list:
        if Levenshtein.distance(i, j) <= 6:
            print(f'{i} -> {j}')
        
```

Постройте модель kNN для распознавания цифр. Посчитайте получившееся значение точности (последняя строка в файле 7. kNN digits.ipynb)

```python
import pandas as pd
```

```python
data_train = pd.read_csv('train_digits.csv')
```

```python
data_train.head()
```

```python
data_test = pd.read_csv('test_digits.csv')
```

```python
data_test.head()
```

```python
data = data_train
```

```python
data.head()
```

```python
data.shape
```

```python
X = data.drop('label', axis=1)
y = data['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
(train_data, test_data, train_dig, test_dig) = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=10)
model.fit(train_data, train_dig)
```

```python
predictions = model.predict(test_data)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(test_dig, predictions))
```
